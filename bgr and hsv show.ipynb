```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
color =cv2.imread('Image1.jpeg') #default , 1 as color image
color =cv2.resize(color, (750,550)) 
#cv2.imshow("Imag", color)
plt.imshow(cv2.cvtColor(color, cv2.COLOR_RGB2BGR))
```

```python
#cv2.moveWindow("imag",0,0)

height,width,channels = color.shape
print(color.shape)
```

```python
b,g,r = cv2.split(color)
print(b.shape)

rgb_split = np.zeros([height,width*3,3],'uint8') 
rgb_split[:,0:width]=cv2.merge([b,b,b])
rgb_split[:,width:2*width]=cv2.merge([r,r,r])
rgb_split[:,2*width:3*width]=cv2.merge([g,g,g])
#cv2.imshow('allcolors',rgb_split)
#cv2.moveWindow('allcolors',height,0)
#print (rgb_split)
```

```python
hsv = cv2.cvtColor(color,cv2.COLOR_BGR2HSV)
h,s,v =cv2.split(hsv)
hsv_split = np.concatenate((h,s,v),axis =1)
cv2.imshow('hsvwalabhai',hsv_split)
plt.imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR))
```
