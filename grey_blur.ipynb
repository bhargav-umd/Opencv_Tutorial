```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Image1.jpeg",1)
img = cv2.resize(img, (750,550))

gray_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)

#cv2.imshow('greybhai',gray_image)

b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]
#rgba = cv2.merge((b,g,r,g))
#cv2.imwrite('transparent1.png',rgba)



```

```python
blur = cv2.GaussianBlur(img,(115,65),1)
#cv2.imshow('blur',blur)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_RGB2BGR))
```

```python
kernel = np.ones((5,5),'uint8')
erode  = cv2.erode(img,kernel,iterations = 1)
dilate = cv2.dilate(img,kernel,iterations =1)

cv2.imshow("dilate",dilate)
cv2.imshow('erode',erode)
plt.imshow(cv2.cvtColor(erode, cv2.COLOR_RGB2BGR))
plt.imshow(cv2.cvtColor(dilate, cv2.COLOR_RGB2BGR))
```
